### Generate some summary statistics for the crime reports data.

**1. Number of crimes of each type**

```python
crime_df["primary_type"].value_counts()
```

**2. How the number of crimes of each type change over time**

```python
crime_df.groupby(["primary_type", "month", "year"]).size().unstack().unstack()
```
